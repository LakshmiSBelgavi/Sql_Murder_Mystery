```python
import requests
from bs4 import BeautifulSoup

# Function to scrape book details from a given page
def scrape_books(page_url):
    response = requests.get(page_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    books = []

    for book in soup.find_all('h3'):
        title = book.a['title']
        rating = book.find_previous('p', class_='star-rating')['class'][1]

        # Modified code to properly extract the price
        price_element = book.find_next('div', class_='product_price').find('p')
        price = price_element.text.strip() if price_element else 'N/A'

        link = 'http://books.toscrape.com/' + book.a['href']

        books.append({'Title': title, 'Rating': rating, 'Price': price, 'Link': link})

    return books

# Function to scrape books from all pages
def scrape_all_books():
    all_books = []

    for page_num in range(1, 51):  # Assuming 50 pages in total
        page_url = f'http://books.toscrape.com/catalogue/page-{page_num}.html'
        all_books.extend(scrape_books(page_url))

    return all_books

# Displaying the output in tabular format (you can use pandas for better table formatting)
def display_books_table(books):
    print("{:<50} {:<10} {:<10} {:<50}".format('Title', 'Rating', 'Price', 'Link'))
    print("="*120)
    for book in books:
        print("{:<50} {:<10} {:<10} {:<50}".format(book['Title'], book['Rating'], book['Price'], book['Link']))

# Scraping and displaying the results
all_books = scrape_all_books()
print(tabulate(all_books,headers='keys',tablefmt='psql'))
```
